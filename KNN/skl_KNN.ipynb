```python
import numpy as np
import pandas as pd

import scipy
import matplotlib.pyplot as plt
# from pylab import reParams
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn import neighbors
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

from sklearn import metrics
```

**Splitting data into test and training datasets**

```python
df = pd.read_csv('/content/drive/MyDrive/dataset-ml/mtcars.csv')
df
```

```python
X_prime = df.iloc[:,[1,3,4,5]].values
y = df.iloc[:,9].values
```

```python
X = preprocessing.scale(X_prime)
```

```python
from pandas.core.common import random_state
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 17)
```

**Building and training model with training data**

```python
clf = neighbors.KNeighborsClassifier()
clf.fit(X_train, y_train)
print(clf)
```

**Evaluating model's prediction against the test dataset**

```python
y_expect = y_test
y_pred = clf.predict(X_test)

print(metrics.classification_report(y_expect, y_pred))
```
